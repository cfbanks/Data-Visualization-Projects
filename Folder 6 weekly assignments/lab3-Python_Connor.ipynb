```python
# YOUR ANSWER HERE
jobs_df = pd.read_csv('cities-job-satisfaction.csv')

ex_chart = alt.Chart(jobs_df, title=alt.TitleParams(text="Dramatic increase in dissatisfaction at work", 
                    subtitle="Can't get no job satisfaction",fontSize=24, align='left', anchor='start', dy=-10, dx=-10, subtitleFontSize=12)
                    ).mark_circle(size=60, opacity = 0.6).encode(
    alt.X('last_year'),
    alt.Y('this_year')).configure_axis(titleFontSize=12, labelFontSize=9, labelPadding=4,
                                       labelColor='grey', titlePadding=10, tickSize=10, tickColor='grey', tickOpacity=0.4, domain=False)
ex_chart
```

```python
jobs_df.rename(columns={'last_year': 'Last Year', 'this_year': 'This Year'}, inplace=True)
```

```python
jobs_df['This Year'].count()
```

```python
# Check for non-realistic job dissatisfaction rates. You can't have a proportion of a population greater than 1.0, so we
# will drop all rows that contain values > 1.0
jobs_df[jobs_df > 1.0].count()
```

We can see that there are 44 entries in the `last year` column that have impossible proportions (> 1.0).
We will drop those rows from the dataframe below. 

```python
# Drop rows with innacurate proportions.
jobs_df.drop(jobs_df[jobs_df['Last Year'] > 1.0].index, inplace=True)
# Confirm proportions have been corrected. 
jobs_df[jobs_df > 1.0].count()
```

```python
#THIS IS THE ONE TO USE, add the other one for information. 

alt.Chart(jobs_df, title=alt.TitleParams(text='Count of US City Job Dissatisfaction Rates')).mark_rect().encode(
    alt.X('Last Year', bin=alt.Bin(maxbins=40), axis=alt.Axis(format='~s')),
    alt.Y('This Year', bin=alt.Bin(maxbins=30)),
    alt.Color('count()'))
```

```python
# Wrangling the data for the plots below so we can plot by year. 
jobs_df.head()
jobs_stacked = pd.DataFrame(jobs_df.stack())
jobs_stacked = jobs_stacked.reset_index()
jobs_stacked.drop('level_0', axis=1, inplace=True)
jobs_stacked.rename(columns= {'level_1': 'Year', 0:'Value'}, inplace=True)
```

```python
stripplot = alt.Chart(jobs_stacked, width=60).mark_circle(size=8).encode(
    x=alt.X(
        'jitter:Q',
        title=None,
        axis=alt.Axis(ticks=True, grid=False, labels=False),
        scale=alt.Scale(),
    ),
    y=alt.Y('Value:Q', title='Job Dissatisfaction Rate'),
    color=alt.Color('Year', legend=None),
    column=alt.Column(
        'Year',
        header=alt.Header(
            labelAngle=0,
            titleOrient='top',
            labelOrient='bottom',
            labelAlign='center',
            labelPadding=10,
            title=None
        ),
    ),
).transform_calculate(jitter='sqrt(-2*log(random()))*cos(2*PI*random())'
).configure_facet(
    spacing=0
).configure_view(
    stroke=None
)
stripplot.title= alt.TitleParams(text=['US City Job', 'Dissatisfaction Rates'], anchor='middle', dx=15)

stripplot
```

```python
 

area = alt.Chart(jobs_stacked).transform_density(
    'Value', groupby=['Year'], as_=['Value', 'density'], steps=200).mark_area(opacity=0.4).encode(
    alt.X('Value', title='Job Dissatisfaction Rate'),
    alt.Y('density:Q', title='Density'),
    alt.Color('Year', legend=alt.Legend(orient='top-right', title=None)))

line = alt.Chart(jobs_stacked).transform_density(
    'Value', groupby=['Year'], as_=['Value', 'density'],
    steps=200).mark_line().encode(
    alt.X('Value'),
    alt.Y('density:Q'),
    alt.Color('Year'))

mean = alt.Chart(jobs_stacked).mark_rule().encode(
    x='mean(Value):Q',
    size=alt.value(2), color='Year')

Density =  area + line + mean
Density.title= alt.TitleParams(text='Density Distribution of Yearly City Job Dissatisfaction Rate',
                              subtitle="Distribution Means Marked with a Vertical Line")

Density
```

```python
# YOUR ANSWER HERE
bad_chart = alt.Chart(gun_df).mark_line(color='black',
                        point=alt.OverlayMarkDef(filled=True, fill='black')).encode(
    alt.X('Year', scale=alt.Scale(zero=False), axis=alt.Axis(grid=False)), 
    alt.Y('Firearm murders', scale=alt.Scale(reverse=True), title='Firearm murders, zero'))

new_bad = bad_chart + bad_chart.mark_area(color='#b41319', opacity=0.8, 
    line=alt.OverlayMarkDef(color='black'), point=alt.OverlayMarkDef(color='black'))

new_text = new_bad + alt.Chart(gun_df.query('Year == 2006')).mark_text(dy=-12, dx=60).encode(
    alt.X('Year'),
    alt.Y('Firearm murders'),
    text=alt.value('Gun law enacted'),
    size=alt.value(16), color=alt.value('white'))

new_text.configure_axis(titleFontWeight=800)
```

```python
# YOUR ANSWER HERE
# Change year to datetime to fix axis labels in chart
gun_df['Year'] = pd.to_datetime(gun_df['Year'], format='%Y')

good_chart = alt.Chart(gun_df, title=alt.TitleParams(text="Firearm Deaths in Florida", fontSize=20)).mark_line(color='black',
                        point=alt.OverlayMarkDef(filled=True, fill='black')).encode(
    alt.X('Year', axis=alt.Axis(grid=False, formatType='time', format='%Y')), 
    alt.Y('Firearm murders', title='Firearm Deaths'))

# Add text to point where gun law enacted. 
new_text = good_chart + alt.Chart(gun_df.query('Year == 2006')).mark_text(dy=5, dx=50).encode(
    alt.X('Year'),
    alt.Y('Firearm murders'),
    text=alt.value('Gun law enacted'),
    size=alt.value(12), color=alt.value('black'))

# Highlight dot where new law was enacted (the first chart says 2005, but yours looks like 2006 so I chose 2006)
new_dot = new_text + alt.Chart(gun_df.query('Year == 2006')).mark_point().encode(
    alt.X('Year'),
    alt.Y('Firearm murders'),
    color=alt.value('red'))

new_dot.configure_axis(titleFontWeight=800, titlePadding=20)
```
